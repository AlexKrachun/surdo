# учусь перебирать видео

```python
train_video_path_prefix = 'slovo/train/'
```

```python
video_names = train_data['attachment_id'] + '.mp4'
```

```python
c = 0
for video_name in video_names:
    with open(train_video_path_prefix + video_name, 'rb') as f:
        print(f.readline()[:5])
        c += 1
print(c)
```
